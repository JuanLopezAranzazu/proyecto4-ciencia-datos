```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
sns.set(style='whitegrid')
```

```python
# Cargar el dataset
df = pd.read_csv("../data/students_stress_level.csv")
```

```python
# Mostrar los primeros 5 registros del DataFrame
print("Primeros 5 registros del DataFrame:")
df.head()
```

```python
# Mostrar información del DataFrame
print("Información del DataFrame:")
df.info()
```

```python
# Mostrar los tipos de datos
print("Tipos de datos en cada columna:")
print(df.dtypes)
```

```python
# Mostrar estadísticas descriptivas generales
print("Estadísticas descriptivas generales:")
df.describe()
```

```python
# Mostrar media y mediana de las características numéricas
print("Media de las características numéricas:")
print(df.select_dtypes(include=[np.number]).mean())

print("\nMediana de las características numéricas:")
print(df.select_dtypes(include=[np.number]).median())
```

```python
# Mostrar la varianza y desviación estándar de las características numéricas
print("Varianza de las características numéricas:")
print(df.select_dtypes(include=[np.number]).var())

print("\nDesviación estándar de las características numéricas:")
print(df.select_dtypes(include=[np.number]).std())
```

```python
# Seleccionar las columnas numéricas
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
print("Columnas numéricas:", numerical_columns)

# Seleccionar las columnas categóricas
categorical_columns = df.select_dtypes(include=['object']).columns
print("Columnas categóricas:", categorical_columns)
```

```python
# Histograma de las variables numéricas
num_vars = len(numerical_columns)
cols = 4  # Número de columnas que deseas
rows = math.ceil(num_vars / cols)

plt.figure(figsize=(5 * cols, 4 * rows))

for i, column in enumerate(numerical_columns, 1):
    plt.subplot(rows, cols, i)
    sns.histplot(df[column], kde=True, bins=30)
    plt.title(f'Histograma de {column}')
    plt.xlabel(column)
    plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

```python
# Boxplots para detectar outliers
num_vars = len(numerical_columns)
cols = 4  # Número de columnas que deseas
rows = math.ceil(num_vars / cols)

plt.figure(figsize=(5 * cols, 4 * rows))

for i, column in enumerate(numerical_columns, 1):
    plt.subplot(rows, cols, i)
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot de {column}')
    plt.xlabel(column)

plt.tight_layout()
plt.show()
```

```python
# Correlación entre variables mediante heatmaps

# Calcular la matriz de correlación
correlation_matrix = df[numerical_columns].corr()

# Crear un heatmap de la matriz de correlación
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matriz de correlación entre variables numéricas')
plt.show()
```

```python
# Calcular la correlación de todas las variables numéricas con stress_level
correlation = df.corr(numeric_only=True)['stress_level'].sort_values(ascending=False)

# Eliminar la correlación de stress_level consigo mismo
correlation = correlation.drop('stress_level')

# Mostrar las 10 más correlacionadas
print("Las 10 variables más correlacionadas con stress_level:")
print(correlation.head(10))

# Gráfico de barras de las 10 variables más correlacionadas con stress_level
plt.figure(figsize=(10, 6))
sns.barplot(x=correlation.head(10).index, y=correlation.head(10).values)
plt.title('Variables más correlacionadas con stress_level')
plt.xlabel('Variables')
plt.ylabel('Correlación')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de barras para las varables numéricas

num_vars = len(numerical_columns)
cols = 4  # Número de columnas que deseas
rows = math.ceil(num_vars / cols)

plt.figure(figsize=(5 * cols, 4 * rows))

for i, column in enumerate(numerical_columns, 1):
    if column != 'stress_level':  # Evitar la columna objetivo
        plt.subplot(rows, cols, i)
        sns.barplot(data=df, x='stress_level', y=column, palette='Set2', hue='stress_level')
        plt.title(f'Gráfico de barras de {column} vs Stress Level')
        plt.xlabel('Stress Level')
        plt.ylabel(column)

plt.tight_layout()
plt.show()
```

```python
# Selección de variables más correlacionadas con 'stress_level'
selected_features = correlation.head(10).index.tolist()

# Graficar las variables seleccionadas
plt.figure(figsize=(15, 10))
for i, feature in enumerate(selected_features, 1):
    plt.subplot(3, 4, i)
    sns.barplot(data=df, x='stress_level', y=feature, palette='Set2', hue='stress_level')
    plt.title(f'{feature} vs Stress Level')
    plt.xlabel('Stress Level')
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```

```python
# Grafica de barras de cada variable selected_features

plt.figure(figsize=(15, 10))

for i, feature in enumerate(selected_features, 1):
    plt.subplot(3, 4, i)
    sns.countplot(data=df, x=feature)
    plt.title(f'Frecuencia de {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

```python
# Histograma de las variables seleccionadas
plt.figure(figsize=(15, 10))

for i, feature in enumerate(selected_features, 1):
    plt.subplot(3, 4, i)
    sns.histplot(df[feature], kde=True, bins=30)
    plt.title(f'Histograma de {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

```python
# Boxplots de las variables seleccionadas
plt.figure(figsize=(15, 10))

for i, feature in enumerate(selected_features, 1):
    plt.subplot(3, 4, i)
    sns.boxplot(x=df[feature])
    plt.title(f'Boxplot de {feature}')
    plt.xlabel(feature)
    
plt.tight_layout()
plt.show()
```
